1. P511 배낭함수 문제의 1, 2에 적합한 코드를 골라서 작성하시오.  

```python
def knapsack():
    array = [[0 for _ in range(maxWeight+1)] for _ in range(rowCount+1)]

    for row in (range(1, rowCount+1)):
        for col in range(1, maxWeight+1):
            if range(weight[row] > col):
                array[row][col] = array[row-1][col]
            else:
                value1 = money[row] + array[row-1][col-weight[row]]
                value2 = array[row-1][col]
                array[row][col] = max(value1, value2)
           
    return array[rowCount][maxWeight]


maxWeight = 7
rowCount = 4
weight = [0, 6, 4, 3, 5]
money = [0, 13, 8, 6, 12]


maxValue = knapsack()
print(f'가방에 담을 수 있는 최대 금액 : {maxValue}')

# 1. range(1, rowCount+1)
# 2. weight[row] > col
```

2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.

```python
## 함수 선언 부분 ##
def isQueueFull():
    global SIZE, queue, front, rear
    if (rear == SIZE-1):
        return True
    else:
        return False
    
def isQueueEmpty():
    global SIZE, queue, front, rear
    if (front == rear):
        return True
    else:
        return False
    
def enQueue(data):
    global SIZE, queue, front, rear
    if (isQueueFull()):
        print('Queue is Full')
        return
    rear += 1
    queue[rear] = data

def deQueue():
    global SIZE, queue,front, rear
    if (isQueueEmpty()):
        print('Queue is Empty')
        return None
    front += 1
    data = queue[front]
    queue[front] = None
    
    for i in range(front+1, rear+1):
        queue[i-1] = queue[i]
        queue[i] = None
    front = -1
    rear -= 1

    return data
    
def peek():
    global SIZE, front, rear, queue
    if (isQueueEmpty()):
        print('큐가 비었습니다.')
        return None
    return queue[front+1]

# 전역 변수 선언
SIZE = int(input('대기줄 입력(숫자) > '))
queue = [None for _ in range(SIZE)]
front = rear = -1

# 메인 코드
if __name__ == '__main__':
    while True:
        select = input('삽입(I)/추출(E)/확인(V)/종료(Q) --> ').upper()

        if ( select == 'Q'):
            print('식당 영업 종료!')
            break

        elif ( select == 'I' ):
             data = input('대기자 이름 입력 > ')
             enQueue(data)
             print(f'대기 줄 상태 : {queue}')

        elif ( select == 'E' ):
            data = deQueue()
            print(f'추출한 데이터 : {data}')
            print(f'대기 줄 상태 : {queue}')

        elif ( select == 'V' ):
            data = peek()
            print(f'확인한 데이터 : {data}')
            print(f'대기 줄 상태 : {queue}')

        else:
            print('다시 입력')
        
```

3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.
데이터의 종류는 [ 콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵 ]

```python
import random

# 함수 선언
class TreeNode():
    def __init__ (self):
        self.left = None
        self.data = None
        self.right = None

def preorder(node):
    if node == None:
        return
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

# 전역 변수 선언
memory = []
root = None
dataAry = ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', '척오리지널 메가 사워',
           '요아정 요거트컵', '페퍼로니피자주먹밥', '널담 슈톨렌', '딸기마시멜로케이크', 
           '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵' ]
print(f'현재 판매중인 물품 목록 : {dataAry}')
# sellAry = [random.choice(dataAry) for _ in range(20)]
sellAry = []
if __name__ == '__main__':
    while True:
        select = input('판매물품 추가(I)/판매완료 물품 입력(E)/판매된 물품 확인(V)/종료(Q) --> ').upper()

        if ( select == 'Q'):
            print('편의점 영업 종료!')
            break

        elif ( select == 'I' ):
            data = input('판매 목록에 추가할 물품 입력 > ')
            dataAry.append(data)
            print(f'추가한 물품 > {data}')
            print(f'현재 판매중인 물품 목록 : {dataAry}')

        elif ( select == 'E' ):
            data = input('판매한 물품 입력 > ')
            for i in range(0, len(dataAry)):
                if dataAry[i] == data:
                    sellAry.append(data)
                    print(f'판매한 물품 : {data}')
                    break
                
                elif i+1 == len(dataAry):
                    print(f'판매목록에 {data}없음')
            print(f'현재까지 판매된 물품 : {sellAry}')

        elif ( select == 'V' ):
            print(f'판매된 물품 확인 > {sellAry}')

        else:
            print('다시 입력하세요.')  
print(f'오늘 판매된 물건(중복O) --> {sellAry}')



# 메인 코드
node = TreeNode()
node.data = sellAry[0]
root = node
memory.append(node)

for name in sellAry[1:]:
        node = TreeNode()
        node.data = name

        current = root
        while True:
            if name == current.data:
                break

            if name < current.data:
                if current.left == None:
                    current.left = node
                    memory.append(node)
                    break
                current = current.left
            else:
                if current. right == None:
                    current.right = node
                    memory.append(node)
                    break
                current = current.right

print(f'오늘 판매된 물건(중복X) --> ', end=' ')
preorder(root)


```

4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.

```python
import random

## 클래스와 함수 선언
def binSearch(ary, data):
    global cnt
    pos = -1
    start = 0
    end = len(ary) - 1 

    while (start <= end):  
        mid = (start + end ) // 2
        cnt += 1
        print(f'검색횟수({cnt}) : arr[{mid}] ==> {ary[mid]}')
        
        if data == ary[mid]:
            return mid
        elif data > ary[mid]:
            start = mid + 1
        else:
            end = mid - 1  
        

    return pos

# 변수 선언
cnt = 0
dataAry = [random.choice(range(0, 100000 + 1)) for _ in range(100000)]
dataAry.sort()

# 메인 로직
print(f'배열 일부 --> {dataAry}')
# print(f'데이터 수 : {len(dataAry)}')

findData = int(input(f'찾을 숫자를 입력하세요(0 ~ 100000) > '))
position = binSearch(dataAry, findData)
print()

if position == -1:
    print(f'{findData}(은)는 없네요.')
else:
    print(f'{findData}(은)는 [{position}] 위치에 있음')
print(f'### {cnt}회 검색 ###')
   
```
